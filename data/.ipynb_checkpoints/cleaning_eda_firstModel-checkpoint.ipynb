```python
import pandas as pd
from datetime import datetime as dt
from datetime import timedelta
import matplotlib.pyplot as plt
import numpy as np
```

```python
transactions = pd.read_csv('transactions.csv')
accounts = pd.read_csv('accounts.csv')
```

```python
transactions['date'] = transactions['date'].apply(lambda x: dt.strptime(x, '%Y-%m-%d'))
accounts['update_date'] = accounts['update_date'].apply(lambda x: dt.strptime(x, '%Y-%m-%d'))

data = pd.merge(transactions, accounts, how='inner', left_on='account_id', right_on='id')
data['time_delta'] = data[['date', 'update_date']].apply(lambda x: (x[1] - x[0]).days, axis=1)
```

```python
def get_ids_enough_history(data, threshold=180):
    seniority = dict(data.groupby('account_id').agg({'time_delta': 'max'}).to_records())
    valid_ids = [key for key in seniority.keys() if seniority[key] >= threshold]
    return valid_ids
```

```python
valid_ids = get_ids_enough_history(data)
```

```python
def get_user_aggregated_data(user_id, data, mode='all', agg_func='sum'):
    if mode == 'outgoing':
        records = dict(data[(data.account_id == user_id) & (data.amount <= 0)].\
                       groupby('date', sort=True).agg({'amount': agg_func}).\
                       to_records())
    elif mode == 'ingoing':
        records = dict(data[(data.account_id == user_id) & (data.amount > 0)].\
                       groupby('date', sort=True).agg({'amount': agg_func}).\
                       to_records())
    else:
        records = dict(data[(data.account_id == user_id)].\
                       groupby('date', sort=True).\
                       agg({'amount': agg_func}).\
                       to_records())
    
    records = {str(key).split('T')[0]: value for key, value in records.items()}
    record_key = list(records.keys())
    start_date = dt.strptime(record_key[0], '%Y-%m-%d')
    end_date = dt.strptime(record_key[-1], '%Y-%m-%d')
    list_dates = [str(start_date + timedelta(days=x)).split(' ')[0] \
                  for x in range((end_date-start_date).days)]
    corrected_records = []
    for day in list_dates:
        if day in record_key:
            corrected_records.append([day, records[day]])
        else:
            corrected_records.append([day, 0])
    
    return dict(corrected_records)

len(get_user_aggregated_data(valid_ids[31], data))
```

```python
# outcomings/incoming for n last months
def calculate_last_amounts(records, n_months=5):
    transactions = list(records.values())
    retVal = []
    n = len(transactions)
    for i in range(1, n_months+1):
        retVal.append(sum(transactions[n-i*30: n-(i-1)*30]))
    return retVal

print(calculate_last_amounts(get_user_aggregated_data(valid_ids[31], data, mode = 'outgoing', agg_func='sum')))        
print(calculate_last_amounts(get_user_aggregated_data(valid_ids[31], data, mode = 'ingoing', agg_func='sum')))        

print(calculate_last_amounts(get_user_aggregated_data(valid_ids[31], data, mode = 'outgoing', agg_func='count')))
print(calculate_last_amounts(get_user_aggregated_data(valid_ids[31], data, mode = 'ingoing', agg_func='count')))
```

```python
class CustomException(Exception):
    def __init__(self, message, errors):
        super().__init__(message)


def get_balance_date(data, account_id):
    
    last_balance = float(data[data.account_id == account_id]['balance'].sample())
    outgoing = get_user_aggregated_data(account_id, data, mode = 'outgoing', agg_func='sum')
    ingoing = get_user_aggregated_data(account_id, data, mode = 'ingoing', agg_func='sum')
    
    ok = list(outgoing.keys())
    ik = list(ingoing.keys())
    
    if len(ok) > 0 and len(ik) > 0:
        min_date = min([dt.strptime(ok[0], '%Y-%m-%d'), dt.strptime(ik[0], '%Y-%m-%d')])
        max_date = max([dt.strptime(ok[-1], '%Y-%m-%d'), dt.strptime(ik[-1], '%Y-%m-%d')])
    elif len(ok) == 0:
        if len(ik) == 0:
            raise CustomException
        else:
            min_date = dt.strptime(ik[0], '%Y-%m-%d')
            max_date = dt.strptime(ik[-1], '%Y-%m-%d')
    elif len(ik) == 0:
        min_date = dt.strptime(ok[0], '%Y-%m-%d')
        max_date = dt.strptime(ok[-1], '%Y-%m-%d')
        
    list_dates = [str(min_date + timedelta(days=x)).split(' ')[0] \
                  for x in range((max_date-min_date).days)]
    outgoing_adapted, ingoing_adapted = [], []
    for date in list_dates:
        if date in outgoing.keys():
            outgoing_adapted.append([date, outgoing[date]])
        else:
            outgoing_adapted.append([date, 0])
        if date in ingoing.keys():
            ingoing_adapted.append([date, ingoing[date]])
        else:
            ingoing_adapted.append([date, 0])
        
    outgoing_adapted, ingoing_adapted = dict(outgoing_adapted), dict(ingoing_adapted)
    
    # check if they have the same period
    balance = {}
    for (date, ingoing), (_, outgoing) in zip(ingoing_adapted.items(), outgoing_adapted.items()):
        current_balance = last_balance + ingoing - abs(outgoing)
        last_balance = current_balance
        balance[date] = current_balance
    return balance

def get_balance_period(data, account_id, n_months=5):
    records = get_balance_date(data, account_id)
    return calculate_last_amounts(records, n_months=n_months)

get_balance_period(data, valid_ids[31])
```

```python
def get_user_aggregated_data(user_id, data, mode='all', agg_func='sum'):
    if mode == 'outgoing':
        records = dict(data[(data.account_id == user_id) & (data.amount <= 0)].\
                       groupby('date', sort=True).agg({'amount': agg_func}).\
                       to_records())
    elif mode == 'ingoing':
        records = dict(data[(data.account_id == user_id) & (data.amount > 0)].\
                       groupby('date', sort=True).agg({'amount': agg_func}).\
                       to_records())
    else:
        records = dict(data[(data.account_id == user_id)].\
                       groupby('date', sort=True).\
                       agg({'amount': agg_func}).\
                       to_records())
    
    if len(records) > 0:
        records = {str(key).split('T')[0]: value for key, value in records.items()}
        record_key = list(records.keys())
        start_date = dt.strptime(record_key[0], '%Y-%m-%d')
        end_date = dt.strptime(record_key[-1], '%Y-%m-%d')
        list_dates = [str(start_date + timedelta(days=x)).split(' ')[0] \
                      for x in range((end_date-start_date).days)]
        corrected_records = []
        for day in list_dates:
            if day in record_key:
                corrected_records.append([day, records[day]])
            else:
                corrected_records.append([day, 0])

        return dict(corrected_records)
    else:
        return dict()


LOOK_BACK_WINDOW = 5 # months    

def feature_engineering(outgoing_by_day, ingoing_by_day, balance_by_day):
    global LOOK_BACK_WINDOW
    ok = list(outgoing_by_day.keys())
    ik = list(ingoing_by_day.keys())
    bk = list(balance_by_day.keys())
    
    # code better
    if len(ok)*len(ik)*len(bk) == 0:
        if len(ok) == 0:
            min_date = min([dt.strptime(ik[0], '%Y-%m-%d'), 
                            dt.strptime(bk[0], '%Y-%m-%d')])
            max_date = max([dt.strptime(ik[-1], '%Y-%m-%d'), 
                            dt.strptime(bk[-1], '%Y-%m-%d')])
        if len(ik) == 0:
            min_date = min([dt.strptime(ok[0], '%Y-%m-%d'), 
                            dt.strptime(bk[0], '%Y-%m-%d')])
            max_date = max([dt.strptime(ok[-1], '%Y-%m-%d'), 
                            dt.strptime(bk[-1], '%Y-%m-%d')])
        if len(bk) == 0:
            min_date = min([dt.strptime(ok[0], '%Y-%m-%d'), 
                            dt.strptime(ik[0], '%Y-%m-%d')])
            max_date = max([dt.strptime(ok[-1], '%Y-%m-%d'), 
                            dt.strptime(ik[-1], '%Y-%m-%d')])
    else:
        min_date = min([dt.strptime(ok[0], '%Y-%m-%d'), 
                        dt.strptime(ik[0], '%Y-%m-%d'), 
                        dt.strptime(bk[0], '%Y-%m-%d')])
        max_date = max([dt.strptime(ok[-1], '%Y-%m-%d'), 
                        dt.strptime(ik[-1], '%Y-%m-%d'), 
                        dt.strptime(bk[-1], '%Y-%m-%d')])
    
    list_dates = [str(min_date + timedelta(days=x)).split(' ')[0] \
                  for x in range((max_date-min_date).days)]
    outgoing_adapted, ingoing_adapted, balance_adapted = [], [], []
    for date in list_dates:
        if date in outgoing_by_day.keys():
            outgoing_adapted.append([date, outgoing_by_day[date]])
        else:
            outgoing_adapted.append([date, 0])
        if date in ingoing_by_day.keys():
            ingoing_adapted.append([date, ingoing_by_day[date]])
        else:
            ingoing_adapted.append([date, 0])
        if date in balance_by_day.keys():
            balance_adapted.append([date, balance_by_day[date]])
        else:
            balance_adapted.append([date, 0])
    
    outgoing_adapted, ingoing_adapted, balance_adapted = dict(outgoing_adapted), dict(ingoing_adapted), dict(balance_adapted)
    
    assert len(outgoing_adapted) == len(ingoing_adapted) == len(balance_adapted)
    
    # Hypothesis : a month is a 30 day period. 
    number_of_possible_shifts = int(len(outgoing_adapted)/30)-6
    
    data = []
    # sum of outgoing/ingoing/balance per month
    for shift in range(number_of_possible_shifts + 1):
        y_outgoing = None
        y_ingoing = None
        y_balance = None        
        outgoings = list(outgoing_adapted.values())
        x_outgoing = []
        ingoings = list(ingoing_adapted.values())
        x_ingoings = []
        balances = list(balance_adapted.values())
        x_balances = []
        n = len(outgoings)
        for i in range(1, LOOK_BACK_WINDOW+2): # in order to also include the 'target' variable's month (the last one)
            start_index = (n-shift) - i*30
            end_index = (n-shift) - (i-1)*30
            if i == 1: # last month
                y_outgoing = sum(outgoings[start_index: end_index])
                y_ingoing = sum(ingoings[start_index: end_index])
                y_balance = sum(balances[start_index: end_index])
            else:
                x_outgoing.append(sum(outgoings[start_index: end_index]))
                x_ingoings.append(sum(ingoings[start_index: end_index]))
                x_balances.append(sum(balances[start_index: end_index]))
        data.append(x_outgoing+x_ingoings+x_balances+[y_outgoing, y_ingoing, y_balance])
    try:
        return pd.DataFrame(data, columns = ['outgoing_'+str(i) for i in range(len(x_outgoing))]+\
                                         ['ingoing_'+str(i) for i in range(len(x_ingoings))]+\
                                         ['balance_'+str(i) for i in range(len(x_balances))]+\
                                          ['y_outgoing', 'y_ingoing', 'y_balance'])
    except:
        return None
    
    
        

def generate_data(data):
    ids = list(data.account_id.unique())
    retVal = pd.DataFrame()
    for account_id in ids:
        try:
            outgoing_by_day = get_user_aggregated_data(account_id, data, mode="outgoing", agg_func="sum")
            ingoing_by_day = get_user_aggregated_data(account_id, data, mode="ingoing", agg_func="sum")
            balance_by_day = get_balance_date(data, account_id)
            df = feature_engineering(outgoing_by_day, ingoing_by_day, balance_by_day)
            if df is not None:
                df['account_id'] = account_id
                retVal = pd.concat([retVal, df])
            else:
                print('account_id ', account_id, ' is untreatable (too much sparse data, not enough transactions)')
        except CustomException:
            print('account_id ', account_id, ' is untreatable!')
    return retVal
    

accounts_to_eliminate = [972, 978, 1012, 1122]
data = data[~data.account_id.isin(accounts_to_eliminate)]
training_data = generate_data(data)
```
